```python
sales = {}
with open("sales.txt") as f:
    for line in f:
        try:
            p,c,price,q = line.strip().split(",")
            total = float(price)*int(q)
            sales[c] = sales.get(c,0)+total
        except:
            continue

print(sales)
```
