```python
transactions = [("A",5000,1),("A",7000,2),("B",2000,10)]
threshold = 6000
flagged = {}

for user, amt, time in transactions:
    if amt > threshold:
        flagged[user] = "High amount"
    for u2, a2, t2 in transactions:
        if user==u2 and abs(time-t2)<=2 and time!=t2:
            flagged[user] = "Frequent transactions"

print(flagged)
```
