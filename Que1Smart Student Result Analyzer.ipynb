```python
try:
    with open("students.txt") as f:
        for line in f:
            try:
                id, name, m1, m2, m3 = line.strip().split(",")
                total = int(m1)+int(m2)+int(m3)
                percent = total/3
                grade = "A" if percent>=80 else "B" if percent>=60 else "C"
                print(name, total, percent, grade)
            except:
                print("Invalid data:", line)
except FileNotFoundError:
    print("File not found")
```

```python
transactions = [("A",5000,1),("A",7000,2),("B",2000,10)]
threshold = 6000
flagged = {}

for user, amt, time in transactions:
    if amt > threshold:
        flagged[user] = "High amount"
    for u2, a2, t2 in transactions:
        if user==u2 and abs(time-t2)<=2 and time!=t2:
            flagged[user] = "Frequent transactions"

print(flagged)
```
